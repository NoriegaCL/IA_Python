```python
#exemplo de manipulaÃ§Ã£o de dados e apresentaÃ§Ã£o de grÃ¡ficos de dispersÃ£o para
#encontrar tendÃªncias.
escolassp = enade.loc[enade['Sigla da UF']=='SP']
escolassp.head()
```

```python
#SerÃ¡ que existe uma relaÃ§Ã£o entre o âNÃºmero de Concluintesâ e a âNota Padronizadaâ?
nconcluintes = escolassp['NÂº  de Concluintes Participantes']
nota = escolassp['Nota Padronizada - CE']
```

```python
plt.scatter(nconcluintes,nota)
plt.show()
```

```python
#Podemos tambÃ©m inverter os valores de x e y:
plt.scatter(nota,nconcluintes)
plt.show()
```

```python
#Por fim vamos apresentar a relaÃ§Ã£o entre nÃºmero de concluintes e nÃºmero de participantes.
nparticipantes = escolassp['NÂº de Concluintes Inscritos']
plt.scatter(nconcluintes,nparticipantes)
plt.show()
```

 parece existir uma relaÃ§Ã£o entre x e y: Quanto maior o nÃºmero de concluintes, maior
o nÃºmero de participantes. Vamos colocar labels para ver


```python
 #Colocar nome no eixo x e y com âxlabelâ e âylabelâ
plt.ylabel('Concluintes')
plt.title('Participantes')
```

```python
# IndicaÃ§Ã£o dos limites dos eixos apresentados no grÃ¡fico com âxlimâ e âylimâ:
plt.xlim(0, 400)
plt.ylim(0, 500)
```

```python
#Grade de apresentaÃ§Ã£o com âgridâ:
plt.grid(True)
```
