```python
#exemplo de manipulação de dados e apresentação de gráficos de dispersão para
#encontrar tendências.
escolassp = enade.loc[enade['Sigla da UF']=='SP']
escolassp.head()
```

```python
#Será que existe uma relação entre o “Número de Concluintes” e a “Nota Padronizada”?
nconcluintes = escolassp['Nº  de Concluintes Participantes']
nota = escolassp['Nota Padronizada - CE']
```

```python
plt.scatter(nconcluintes,nota)
plt.show()
```

```python
#Podemos também inverter os valores de x e y:
plt.scatter(nota,nconcluintes)
plt.show()
```

```python
#Por fim vamos apresentar a relação entre número de concluintes e número de participantes.
nparticipantes = escolassp['Nº de Concluintes Inscritos']
plt.scatter(nconcluintes,nparticipantes)
plt.show()
```

 parece existir uma relação entre x e y: Quanto maior o número de concluintes, maior
o número de participantes. Vamos colocar labels para ver


```python
 #Colocar nome no eixo x e y com “xlabel” e “ylabel”
plt.ylabel('Concluintes')
plt.title('Participantes')
```

```python
# Indicação dos limites dos eixos apresentados no gráfico com “xlim” e “ylim”:
plt.xlim(0, 400)
plt.ylim(0, 500)
```

```python
#Grade de apresentação com “grid”:
plt.grid(True)
```

```python
plt.ylabel('Concluintes')
plt.title('Participantes')
plt.xlim(0, 400)
plt.ylim(0, 500)
plt.grid(True)
plt.scatter(nconcluintes,nparticipantes)
plt.show()
```
